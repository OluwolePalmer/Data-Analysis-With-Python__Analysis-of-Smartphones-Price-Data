See the data source file for information about the dataset.

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('smartphones.csv')
df
```

```python
#Identify all the brands of smartphones present in the data.
print('There are',len(df['Brand'].unique()),'brands of smartphones in the data, and they are the following:\n\n',df['Brand'].unique())
```

```python
# Identify the top 10 brands with most smartphones.
df['Brand'].value_counts().head(10).to_frame()
```

```python
# Top 10 brands visualization.
df['Brand'].value_counts().head(10).plot(kind='bar', color='pink')
plt.plot(df['Brand'].value_counts().head(10), color='black', marker='o')
plt.title('Top 10 Smartphone Brands')
plt.ylabel('count')
plt.ylim(0,)
```

```python
# For each brand, determine the count, average price, lowest price, highest price of smartphones using groupby.
df.groupby('Brand', as_index=False).agg({'Smartphone':'count', 'Final Price':['min', 'max', 'mean']})
```

```python
# Using get_group, show details of the brand with the most smartphones.
df.groupby('Brand').get_group('Samsung')
```
